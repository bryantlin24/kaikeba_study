# 1. 从fundemantals.csv开始！

fundemantals.csv 是这些股票的年报数据

请用数据回答以下问题：

1. S&P500股票在2015年`net income`的均值是多少？最大值比最小值多多少？（每问10分，共计20分）
2. S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票是的代码（ticker symbol）是什么？（每问10分，共计20分）


```python
#1
import pandas as pd
df=pd.read_csv('fundamentals.csv')
df= df.rename(columns={'For Year':"For_Year"})
df1=df[df['For_Year']==2015]

print('2015年income的均值是',df1['Net Income'].mean())
print('最大值比最小值多',df1['Net Income'].max()-df1['Net Income'].min())
```

```python
fd_2016 = df.query('For_Year == 2016.0')
fd_2016['Fixed/Total'] = (fd_2016["Fixed Assets"]/fd_2016["Total Assets"])
print('均值是',fd_2016['Fixed/Total'].mean())
min = fd_2016.sort_values(by="Fixed/Total").iloc[0]
name = min["Ticker Symbol"]
print("定资产占总资产比例最小的股票的代码（ticker symbol）是:"+name)
```

# 2. 加入securities.csv~

securities.csv包含了这些股票的基本信息

1. 请列举出各个sector中的加入时间最早的股票名称（10分）
2. 请列举出每一个州中加入时间最晚的股票名称（10分）


```python
df = pd.read_csv('securities.csv',encoding='ANSI')
sectorlist=set(df['GICS Sector'].values.tolist())
earliest={}
for i in sectorlist:
    sectorsecurities=df[df['GICS Sector']==i]
    sectorlisted=sectorsecurities.sort_values(by=['Date first added'])
    first=sectorlisted[sectorlisted['Date first added']==sectorlisted['Date first added'].values[0]]
    earliest[i]=first['Security'].values.tolist()
for i,k in earliest.items():
    print(i,'板块最早加入的股票是',k)
print("***********************************")
    
    
    
statelist=df['Address of Headquarters'].values.tolist()
state=[]
for i in statelist:
    statename=i.split(', ')[1]
    if '[' in statename:
        statename=statename[:statename.find('[')]
    state.append(statename)
state=set(state)
latest={}
for k in state:
    count=[]
    for i in range(0,len(df['Address of Headquarters'])):
        if k in df['Address of Headquarters'].values[i]:
            count.append(i)
    stateall=df.iloc[count]
    stateall=stateall.sort_values(by=['Date first added'],ascending=False)
    last=stateall[stateall['Date first added']==stateall['Date first added'].values[0]]
    latest[k]=last['Security'].values.tolist()
for i,k in latest.items():
    print(i,'最晚的股票是',k)
```

# 3. merge!

现在你需要同时处理来自两个表中的信息了

1. 请思考，合并两个表的信息的时候，我们应该用什么样的准则对其它们（10分）
2. 请列举每个sector在2013-2016年累计Research&Development的总投入（10分）
3. 请列举出每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值（20分）

```python
import pandas as pd
df1 = pd.read_csv('fundamentals.csv',encoding='ANSI')
df2 = pd.read_csv('securities.csv',encoding='ANSI')
df3=pd.merge(df1, df2,how='left',left_on='Ticker Symbol',right_on="Ticker symbol")

count=[]
for i in range(0,len(df3['Research and Development'])):
        if '2013' in df3['Period Ending'].values[i] or '2014' in df3['Period Ending'].values[i] or '2015' in df3['Period Ending'].values[i] or '2016' in df3['Period Ending'].values[i]:
            count.append(i)
df4=df3.iloc[count]
df5=df4['Research and Development'].groupby(df4['GICS Sector']).sum()
for i in range(len(df5)):
    print('{}总投入为{}'.format(df5.index[i],df5.values[i]))
print("***********************************\n")

sectorlist=set(df3['GICS Sector'].values.tolist())
for i in sectorlist:
    df6=df4[df4['GICS Sector']==i]
    df7=df6['Research and Development'].groupby(df6['Ticker Symbol']).sum()
    count1=[]
    for j in range(len(df7)):
        count1.append([df7.index[j],df7.values[j]])
    count1.sort(key=lambda x:x[1],reverse=True)
    if count1[0][1]==0:
        print('{}中所有公司投入为0'.format(i))
        continue
    for k in range(0,3):
        print('{}中投入第{}的公司为{}，投入金额为{}'.format(i,k+1,count1[k][0],count1[k][1]))
```

请导入price.csv，然后结合你的聪明才智回答以下问题（附加题，40分）

假设你是某基金公司的老板，现在对于每只股票，你都专门安排了一位负责它的交易员。公司规定每一位交易员手中的资金要么全部买入要么全部卖出（空仓，转化为现金）。假设2016年每一位交易员手中都有10000美元，假设他们都能够看到2016年全年的数据，假设他们都能抓住每一次机会，那么请问2016年底时，赚钱最多的股票是哪一只，赚了多少钱？

```python
prices = pd.read_csv("prices.csv")
prices["value"] = prices["close"]-prices["open"]
price_group = prices.groupby(by="symbol")
maxs ={}
for name,item in price_group:
    temp = item.query("value>0")
    value = temp["value"].sum()
    maxs[name] = value
    print(name+"______"+str(value))
sa = max(zip(maxs.values(),maxs.keys()))
print(sa[0])
print(sa[1])
```
