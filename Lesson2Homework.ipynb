## 1. 实现一周的记账功能
首先记录输入的7天的收入，然后记录输入的7天的支出。  
然后打印出以下内容：
- 7天的收入(list)
- 7天的支出(list)
- 每天的结余(dict)
- 最终的结余

````python
income = []
output = []
surplus = dict()
final_surplus = 0
for i in range(7):
    perday_income = input('周'+str(i+1)+'的收入：')
    perday_output = input('周'+str(i+1)+'的支出：')
    if int(perday_income) >= 0 and int(perday_output)>= 0:
        income.append(perday_income)
        output.append(perday_output)
        surplus[i+1] = int(income[i]) -int(output[i])
        final_surplus = surplus[i+1] + final_surplus
        print('周'+str(i+1)+'的结余',final_surplus)
    else:
        break
print('七天的收入',income)
print('七天的支出',output)
print('每天的结余',surplus)
print('最终的结余',final_surplus)
````

## 2. 实现ATM机器
- 查询余额
- 存款
- 取款
- 退出

````python
balance =0
while (True):
    print('请输入：1.查询余额  2.存款  3.取款  4.退出')
    button = int(input())
    if button == 1:
        print('余额为',balance)
    elif button == 2:
        income = int(input('输入存款金额：'))
        balance = balance + income
    elif button == 3:
        output = int(input('输入取款金额：'))
        balance = balance - output
        if balance < 0:
            print('余额不足，操作失败')
            balance = balance + output
    elif button == 4:
        print('退出')
        break
        
````
